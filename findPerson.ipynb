```python
import cv2
import matplotlib.pyplot as plt
import cvlib as cv
from cvlib.object_detection import draw_bbox
```

```python
img = cv2.imread("/content/people.jpg")
bbox, label, conf = cv.detect_common_objects(img)
reslut = draw_bbox(img, bbox, label, conf)

print(f"Number of Person: {label.count('person')}")
plt.imshow(reslut)
plt.show()
```
